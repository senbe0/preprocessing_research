```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# CSVファイルを読み込む
df = pd.read_csv("/home/eiji/project/g/data/TON_iot/Processed_IoT_dataset/IoT_Thermostat.csv")
df
```

```python

# OneHotEncoderのインスタンスを作成
encoder = OneHotEncoder(sparse=False)

# ワンホットエンコーディングを適用
encoded = encoder.fit_transform(df[['thermostat_status']])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
# 結果の確認
print(encoded_df.head())

df_combined = pd.concat([df, encoded_df], axis=1)
```

# 時間

```python
df_combined['time'] = df_combined['time'].str.strip()
```

```python
df_combined['time']  = pd.to_datetime(df_combined['time'], format='%H:%M:%S').dt.time
```

```python
df = df_combined
```

```python
df['hour'] = df['time'].apply(lambda x: x.hour)
df['minute'] = df['time'].apply(lambda x: x.minute)
df['second'] = df['time'].apply(lambda x: x.second)
```

```python
df = df.drop(columns=['date', 'time', 'thermostat_status', 'type'])
```

```python
new_order = ['hour', 'minute', 'second', 'current_temperature', 'thermostat_status_0', 'thermostat_status_1', 'label']
```

```python
df = df[new_order]
```
