```python
import pandas as pd

# CSVファイルを読み込む
df = pd.read_csv("/home/eiji/project/g/data/TON_iot/Linux_data/Linux_process_1.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["CMD_encoded"] = encoder.fit_transform(df["CMD"])

df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# OneHotEncoderのインスタンスを作成
encoder = OneHotEncoder(sparse_output=False)

# CMD列のエンコーディングを適用
CMD_encoded = encoder.fit_transform(df[['POLI']])

# エンコードされたデータを DataFrame に変換
CMD_encoded_df = pd.DataFrame(CMD_encoded, columns=encoder.get_feature_names_out(['POLI']), index=df.index)

# ワンホットエンコードされたCMD列を元のDataFrameに結合
df = pd.concat([df.drop(columns=['POLI']), CMD_encoded_df], axis=1)

# 結果を確認
print(df.head())
```
