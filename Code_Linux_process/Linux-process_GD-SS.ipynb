```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# CSVファイルを読み込む
df = pd.read_csv("../preprocessed_data/Linux_process_GD.csv")
df
```

```python
from sklearn.preprocessing import StandardScaler
# 標準化を適用する列
columns_to_scale = ['PID', 'TRUN', 'TSLPI', 'TSLPU', 'NICE', 'PRI', 'RTPR', 'CPUNR', 'EXC', 'CPU']

# StandardScalerのインスタンスを作成
scaler = StandardScaler()

# 標準化の適用
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# 結果を表示
df
```

```python
df.to_csv('../preprocessed_data/Linux_process_GD-SS.csv', index=False)
```
